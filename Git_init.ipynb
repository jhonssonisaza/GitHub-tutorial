# **Git init**
Later of open your folder proyect you need to become it in a local repository.
For it you need open your Terminal with:

Ctrl + ñ

And write **git init** to vinculate.

## **Add files**
Write **git add .** on the terminal and this add all the files.

## **Git commit**
Write **git commit -m "commit name"** and this save a control point where you can see the changes.

# **GitHub connect**
Use your URL such as https://github.com/your-user/my-first-proyect.git and use the funtion on the terminal:

**git branch -M main**

**git remote add origin URL**

## **Uploand the file to Git repository**
Write on the Terminal **git push -u origin main**

# **VERY IMPORTAN**
Before to upload the file you have sure to uptade the changes of your file with **Ctrl + S** (Save your changes).

For every upload you need to realize this:

**git add .**

**git commit -m "Actualice commit version"**

**git push**
